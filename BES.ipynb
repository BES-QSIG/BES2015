```python
%matplotlib inline 
# allows for plots in the notebook format 
    ### next - need to see what is happening at each timestep
#import time
import numpy # useful functions for dealing with numbers and arrays

import matplotlib.pyplot as plt # functions for graphs - making dispalys easier - the 'as plt' means we only need to refer to it as plt in the future

from random import randint

def makeFig():
    plt.imshow(env)
    
plt.ion() # enable interactivity
fig=plt.figure() # make a figure

env_size = 100
shark_no = 10
fish_no = 100
z_plank_no = 1000
p_plank_no = 100

env = numpy.zeros((env_size,env_size))  # creates an environment of size 1:100 x 1:100

#print(env)

shark_x = numpy.round(numpy.random.rand(shark_no) * (env_size-2) + 1) # generate x coords for each shark between 1 and 100  -to fit in environment
shark_y = numpy.round(numpy.random.rand(shark_no) * (env_size-2) + 1)
shark_energy = numpy.zeros((shark_no))
shark_energy[:] = 10

fish_x = numpy.round(numpy.random.rand(fish_no) * (env_size-2) + 1) # generate x coords for each fish between 1 and 100  -to fit in environment
fish_y = numpy.round(numpy.random.rand(fish_no) * (env_size-2) + 1)
fish_energy = numpy.zeros((fish_no))
fish_energy[:] = 10

z_plank_x = numpy.round(numpy.random.rand(z_plank_no) * (env_size-2) + 1) # generate x coords for each zooplank between 1 and 100  -to fit in environment
z_plank_y = numpy.round(numpy.random.rand(z_plank_no) * (env_size-2) + 1) 
z_plank_energy = numpy.zeros((z_plank_no))
z_plank_energy[:] = 10

p_plank_x = numpy.round(numpy.random.rand(p_plank_no) * (env_size-2)) # generate x coords for each phytoplank between 1 and 100  -to fit in environment
p_plank_y = numpy.round(numpy.random.rand(p_plank_no) * (env_size-2))

for x in range(0,p_plank_no-1):
    env[p_plank_x[x],p_plank_y[x]] = env[p_plank_x[x],p_plank_y[x]] + 1
       

#print(shark_x)  # prints out the values in the console so we can actually see if it has worked
#print(shark_y)

for x in range(0,10): # timestep for the entire model
    #P_plank_growth
    for c in range(2,env_size*env_size):
        env2 = env
        a = randint(2,98)
        b = randint(2,98)
        env[a+1,b] = env[a+1,b] + env[a,b] 
        env[a+1,b+1] = env[a+1,b+1] + env[a,b] 
        env[a,b+1] = env[a,b+1] + env[a,b] 
        env[a-1,b+1] = env[a-1,b+1] + env[a,b] 
        env[a-1,b] = env[a+1,b] + env[a,b] 
        env[a-1,b-1] = env[a-1,b-1] + env[a,b] 
        env[a,b-1] = env[a,b-1] + env[a,b]
        env[a+1,b-1] = env[a+1,b-1] + env[a,b]
        env[a,b] = env[a,b] * 2
        #plt.imshow(env) # problem - this plots every single time - uses lots of memory - don't do this... 
      

    # add z_plank eating p_plank but do we want to add dispersal too?    
    # How about add point growth, but dispersal if > certasin value?

    for d in range(0,z_plank_no):
        if(env[z_plank_x[d],z_plank_y[d]] != 0):
            z_plank_energy[d] = z_plank_energy[d] + env[z_plank_x[d],z_plank_y[d]] # working on current environment, previous to growth
            env[z_plank_x[d],z_plank_y[d]] = 0
        else:
            z_plank_energy[d] =  int(z_plank_energy[d]/2)  ## note - death = decomposition
        
        if(z_plank_energy[d] > 20):
            z_plank_energy[d] = 10
            z_plank_no = z_plank_no + 1
            randx = randint(0,2) - 1
         
            randy = randint(0,2) - 1
           
        
            z_plank_x = numpy.hstack((z_plank_x, z_plank_x[d] + randx)) 
            if z_plank_x[z_plank_no-1] > 99:
                z_plank_x[z_plank_no-1] = 99
            if z_plank_x[z_plank_no-1] < 0:
                z_plank_x[z_plank_no-1] = 0
            z_plank_y = numpy.hstack((z_plank_y, z_plank_y[d] + randy))
            if z_plank_y[z_plank_no-1] > 99:
                z_plank_y[z_plank_no-1] = 99
            if z_plank_y[z_plank_no-1] < 0:
                z_plank_y[z_plank_no-1] = 0
            z_plank_energy = numpy.hstack((z_plank_energy, 10))
        
        if(z_plank_energy[d] < 1):
            dd = numpy.array(d) # because reidiculously you can't remove a single element, only an array of elements....
            z_plank_no = z_plank_no - 1
            z_plank_x2 = numpy.delete(z_plank_x, dd)
            z_plank_x = z_plank_x2
            z_plank_y2 = numpy.delete(z_plank_y, dd)
            z_plank_y = z_plank_y2
            z_plank_energy2 = numpy.delete(z_plank_energy, dd)
            z_plank_energy = z_plank_energy2
            
    #print(x)        
    for d in range(0,fish_no): # feeding by fish
        if d == (fish_no-1) or d == fish_no:
            #print('break')
            break
            
        fish_energy[d] = fish_energy[d]-1
        if fish_energy[d] < 1:
            dd = numpy.array(d) # because reidiculously you can't remove a single element, only an array of elements....
            fish_no = fish_no - 1
            fish_x2 = numpy.delete(fish_x, dd)
            fish_x = fish_x2
            fish_y2 = numpy.delete(fish_y, dd)
            fish_y = fish_y2
            fish_energy2 = numpy.delete(fish_energy, dd)
            fish_energy = fish_energy2
        
        for e in range(0,z_plank_no):
            if z_plank_x[e] == fish_x[d] and z_plank_y[e] == fish_y[d]:
                fish_energy[d] = fish_energy[d] + round(0.1*z_plank_energy[e]) ### FLOAT!!! might be a better way to do this
                z_plank_energy[e] = 1 # easier than it dying
            
            
        if fish_energy[d] > 20:
            fish_energy[d] = 10
            fish_no = fish_no + 1
            randx = randint(0,2) - 1
         
            randy = randint(0,2) - 1
           
        
            fish_x = numpy.hstack((fish_x, fish_x[d] + randx)) 
            if fish_x[fish_no-1] > 99:
                fish_x[fish_no-1] = 99
            if fish_x[fish_no-1] < 0:
                fish_x[fish_no-1] = 0
            fish_y = numpy.hstack((fish_y, fish_y[d] + randy))
            if fish_y[fish_no-1] > 99:
                fish_y[fish_no-1] = 99
            if fish_y[fish_no-1] < 0:
                fish_y[fish_no-1] = 0
            fish_energy = numpy.hstack((fish_energy, 10))
            
    for d in range(0,shark_no): # feeding by sharks
        if d == (shark_no-1) or d == shark_no:
            #print('break')
            break
            
        shark_energy[d] = shark_energy[d]-1
        if shark_energy[d] < 1:
            dd = numpy.array(d) # because reidiculously you can't remove a single element, only an array of elements....
            shark_no = shark_no - 1
            shark_x2 = numpy.delete(shark_x, dd)
            shark_x = shark_x2
            shark_y2 = numpy.delete(shark_y, dd)
            shark_y = shark_y2
            shark_energy2 = numpy.delete(shark_energy, dd)
            shark_energy = shark_energy2    
        
    
        
        for e in range(0,fish_no):
            if fish_x[e] == shark_x[d] and fish_y[e] == shark_y[d]:
                shark_energy[d] = shark_energy[d] + round(0.1*fish_energy[e]) ### FLOAT!!! might be a better way to do this
                fish_energy[e] = 1 # easier than it dying 
            
        if shark_energy[d] > 20:
            shark_energy[d] = 10
            shark_no = shark_no + 1
            randx = randint(0,2) - 1
         
            randy = randint(0,2) - 1
           
        
            shark_x = numpy.hstack((shark_x, shark_x[d] + randx)) 
            if shark_x[shark_no-1] > 99:
                shark_x[shark_no-1] = 99
            if shark_x[shark_no-1] < 0:
                shark_x[shark_no-1] = 0
            shark_y = numpy.hstack((shark_y, shark_y[d] + randy))
            if shark_y[shark_no-1] > 99:
                shark_y[shark_no-1] = 99
            if shark_y[shark_no-1] < 0:
                shark_y[shark_no-1] = 0
            shark_energy = numpy.hstack((shark_energy, 10))
            
            
    for y in range(0,len(fish_x)): # movement for fish
      
        direction = randint(0,7)
        if direction == 0:
            fish_y[y] == fish_y[y]-1
        if direction == 1:
            fish_y[y] == fish_y[y]-1
            fish_x[y] == fish_x[y] + 1
        if direction == 2:
            fish_x[y] == fish_x[y] + 1
        if direction == 3:
            fish_y[y] == fish_y[y] + 1
            fish_x[y] == fish_x[y] + 1
        if direction == 4:
            fish_y[y] == fish_y[y] + 1
            #fish_x[y] == fish_x[y] + 1
        if direction == 5:
            fish_y[y] == fish_y[y] + 1
            fish_x[y] == fish_x[y] - 1
        if direction == 6:
            #fish_y[y] == fish_y[y] + 1
            fish_x[y] == fish_x[y] - 1
        if direction == 7:
            fish_y[y] == fish_y[y] - 1
            fish_x[y] == fish_x[y] - 1
        if fish_y[y] >= (env_size-1):
            fish_y[y] = 1
        if fish_y[y] < 1:
            fish_y[y] = (env_size-1)
        if fish_x[y] >= (env_size-1):
            fish_x[y] = 1
        if fish_x[y] < 1:
            fish_x[y] = (env_size-1)    

    
    for y in range(0,len(shark_x)): # movement for sharks
     
        direction = randint(0,7)
        if direction == 0:
            shark_y[y] == shark_y[y]-1
        if direction == 1:
            shark_y[y] == shark_y[y]-1
            shark_x[y] == shark_x[y] + 1
        if direction == 2:
            shark_x[y] == shark_x[y] + 1
        if direction == 3:
            shark_y[y] == shark_y[y] + 1
            shark_x[y] == shark_x[y] + 1
        if direction == 4:
            shark_y[y] == shark_y[y] + 1
            #shark_x[y] == shark_x[y] + 1
        if direction == 5:
            shark_y[y] == shark_y[y] + 1
            shark_x[y] == shark_x[y] - 1
        if direction == 6:
            #shark_y[y] == shark_y[y] + 1
            shark_x[y] == shark_x[y] - 1
        if direction == 7:
            shark_y[y] == shark_y[y] - 1
            shark_x[y] == shark_x[y] - 1
        if shark_y[y] >= (env_size-1):
            shark_y[y] = 1
        if shark_y[y] < 1:
            shark_y[y] = (env_size-1)
        if shark_x[y] >= (env_size-1):
            shark_x[y] = 1
        if shark_x[y] < 1:
            shark_x[y] = (env_size-1)

    makeFig()      

    plt.pause(0.1)
   
   
   
```
